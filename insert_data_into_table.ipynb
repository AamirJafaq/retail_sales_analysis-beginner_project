```python
import pandas as pd
from sqlalchemy import create_engine
```

```python

engine = create_engine(
    "postgresql+psycopg2://postgres:root@localhost:5432/RetailSalesAnalysis"
)
```

```python

# Read CSV
data = pd.read_csv("transactions.csv")


# Write to PostgreSQL
data.to_sql(
    'retail_sales',
    engine,
    schema='public',
    if_exists='append',
    index=False
)

print("Data inserted successfully")
```
